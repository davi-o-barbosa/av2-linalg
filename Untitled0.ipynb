```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
# Importando a imagem
img = cv2.imread('teste.png')

# Taxa de compressão, em %. 
# 80 significa que: de uma imagem de 100px de altura, apenas os 80 primeiros valores (80% de 100px) serão consideradas para a computação.
TAXA_DE_COMPRESSAO = 80

# Obtendo a altura da imagem
altura = img.shape[0]
print(f'Altura: {altura}')

# Calculando o valor do threshold
threshold = round(altura * (TAXA_DE_COMPRESSAO/100))
print(f'Threshold: {threshold}')
```

```python
def comprimir(array, threshold):
  # Separando os canais de cor
  # A imagem pelo cv2 tem três canais, vermelho, verde e azul;

  # A primeira coluna representa o canal azul.
  blue = array[:, :, 0]
  green = array[:, :, 1]
  red = array[:, :, 2]

  # Decompondo as arrays utilizando SVD:
  # O primeiro valor retornado é uma array dos autovetores de esquerda
  # O segundo valor é uma array dos valores sigma, sendo eles os valores singulares
  # O terceiro valor é uma array dos autovetores de direita
  redU, redS, redV = np.linalg.svd(red)
  greenU, greenS, greenV = np.linalg.svd(green)
  blueU, blueS, blueV = np.linalg.svd(blue)

  # Multiplicar as três arrays da função SVD retorna a própria array original se você considerar todas os valores dos três vetores.
  # As arrays retornadas pela função SVD são ordenadas por quão bem cada valor representa a array original.
  # Logo, o valor de sigma (segunda array, tipo redS, greenS e blueS) é decrescente:
  # S[0] > S[1] > S[2] > S[3] ... S[m], sendo m o último valor diferente de zero da array, que é justamente a altura da imagem.
  # Sigma é diferente de zero enquanto m < altura da imagem.

  # Concluindo, as informações mais importantes para calcular a array original estão contidas nos primeiros valores de cada uma
  # das três arrays, então podemos definir um "threshold" de quanta informação queremos manter para realizar este calculo

  # Se tivermos uma imagem de 100px de altura e definirmos um threshold de 50, o algoritmo irá calcular a array utilizando apenas
  # os 50 primeiros valores de U, S e V, perdendo parte da informação. Porém como a array está organizada com os valores mais relevantes no topo,
  # ainda é possível realizar uma reconstrução decente da imagem.

  # multi_dot calcula o produto das três arrays, sendo elas filtradas pelo threshold definido
  # np.diag cria uma array diagonal com os valores do vetor S para o produto ser calculado.

  red = np.linalg.multi_dot([redU[:, :threshold], np.diag(redS[:threshold]), redV[:threshold, :]])
  green = np.linalg.multi_dot([greenU[:, :threshold], np.diag(greenS[:threshold]), greenV[:threshold, :]])
  blue = np.linalg.multi_dot([blueU[:, :threshold], np.diag(blueS[:threshold]), blueV[:threshold, :]])

  # Recriando a estrutura da array original com os canais de cor
  array = np.zeros(array.shape)
  array[:, :, 0] = blue
  array[:, :, 1] = green
  array[:, :, 2] = red
  
  return array
```

**Compressão com threshold em 80%**

```python
print(f'Threshold utilizado: {threshold}\n')
cv2_imshow(comprimir(img, threshold))
```
